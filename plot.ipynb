```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Test": ["testALThard", "testHD", "testPQ", "badtest", "testALTsoft", "testHDmanPQtrace"],
    "Average Bitrate": [500000.0, 500000.0, 50000.0, 500000.0, 500000.0, 50000.0],
    "Buffer Time": [1.001, 0.202, 246.381, 1.012, 0.202, 242.584],
    "Switches": [0, 0, 0, 0, 0, 0],
    "Score": [474975.6363100182, 494846.12366372766, 0.16236394651992062, 474707.7181838023, 494846.12366372766, 0.19727546227026632]
}
df = pd.DataFrame(data)
```

```python
# Plot 1: Average Bitrate per Test
plt.figure(figsize=(12, 3))
plt.bar(df["Test"], df["Average Bitrate"], color="skyblue")
plt.title("Average Bitrate per Test")
plt.xlabel("Test")
plt.ylabel("Average Bitrate (bps)")
plt.show()
```

```python
# Plot 2: Buffer Time per Test
plt.figure(figsize=(12, 3))
plt.bar(df["Test"], df["Buffer Time"], color="salmon")
plt.title("Buffer Time per Test")
plt.xlabel("Test")
plt.ylabel("Buffer Time (seconds)")
plt.show()
```

```python
# Plot 3: Score per Test
plt.figure(figsize=(12, 3))
plt.bar(df["Test"], df["Score"], color="lightgreen")
plt.title("Score per Test")
plt.xlabel("Test")
plt.ylabel("Score")
plt.show()
```

```python
# Plot 4: Comparison of Average Bitrate, Buffer Time, and Score
fig, ax1 = plt.subplots(figsize=(15, 5))

ax1.bar(df["Test"], df["Average Bitrate"], color="skyblue", label="Average Bitrate", width=0.4, align="center")
ax1.set_xlabel("Test")
ax1.set_ylabel("Average Bitrate (bps) / Score", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")
ax1.plot(df["Test"], df["Score"], color="green", marker="o", linestyle="--", label="Score")

ax2 = ax1.twinx()
ax2.plot(df["Test"], df["Buffer Time"], color="red", marker="s", linestyle="--", label="Buffer Time (seconds)")
ax2.set_ylabel("Buffer Time (seconds)", color="red")
ax2.tick_params(axis='y', labelcolor="red")
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))

plt.title("Comparison of Average Bitrate, Buffer Time, and Score per Test")
plt.xticks(rotation=45)
plt.show()
```
